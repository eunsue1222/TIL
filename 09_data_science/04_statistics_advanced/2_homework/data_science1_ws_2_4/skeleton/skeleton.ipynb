```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 포아송분포 데이터 불러오기
df = pd.read_csv('../data/data.csv')
df
```

```python
poisson_samples = df['poisson_samples']
poisson_samples
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(poisson_samples, bins=30, kde=True, edgecolor='black')  # [빈칸] 포아송분포 히스토그램 시각화 코드 작성
plt.title('Data generated with Poisson distribution')
plt.xlabel('Number of vehicles arriving within 1 hour')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

- 포아송 분포(Poisson distribution): 어떤 사건이 일정 시간 혹은 공간 내에 몇 번 발생하는지
  - 독립성
  - 고정된 평균 발생률
  - 희소성
  - 독립적 사건 발생

```python
# 정규분포 근사를 위한 평균과 표준편차 계산
lam = 20
num_samples = 1000

mu = np.mean(poisson_samples)  # 기대값(mean) 계산 코드 작성
sigma = np.sqrt(np.var(poisson_samples))  # 표준편차(std) 계산 코드 작성

# 정규분포 생성
normal_samples = np.random.normal(mu, sigma, num_samples)  # 정규분포 샘플 생성 코드 작성

# 포아송분포와 정규분포 비교
plt.figure(figsize=(8, 5))
sns.histplot(poisson_samples, bins=30, kde=True, color='blue', label='Poisson')  # [빈칸] 포아송분포 시각화 코드 작성
sns.histplot(normal_samples, bins=30, kde=True, color='red', label='Normal', alpha=0.6)  # [빈칸] 정규분포 시각화 코드 작성
plt.title('Comparison of Poisson distribution and normal distribution')
plt.xlabel('Number of vehicles arriving within 1 hour')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

- λ가 충분히 크면(현재: 20), 포아송 분포가 정규 분포와 거의 비슷한 모양으로 근사되어 통계 분석 시 정규 분포를 대신 사용할 수 있다.
